```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style for plots
sns.set(style="whitegrid")
```

```python
df=pd.read_csv("happinessdataset.csv")
df
```

```python
print(df['Country'])
```

```python
#show all rows of the dataset without hiding or shortening the output
pd.set_option('display.max_rows', None)

print(df['Country'])
```

```python
# First 5 rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Statistical summary
df.describe()
```

```python
#check missing values
df.isnull().sum()
```

```python
#Top 10 Happiest Countries
top10 = df.sort_values(by='Happiness Score', ascending=False).head(10)

print(top10[['Country', 'Happiness Score']])
```

```python
#Plot Top 10 Happiest Countries
plt.figure(figsize=(10,6))

sns.barplot(
    x='Happiness Score',
    y='Country',
    data=top10,
    palette='viridis'
)

plt.title('Top 10 Happiest Countries')
plt.xlabel('Happiness Score')
plt.ylabel('Country')

plt.show()
```

```python
#Bottom 10 Countries
bottom10 = df.sort_values(by='Happiness Score').head(10)

print(bottom10[['Country', 'Happiness Score']])
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

```python
#GDP vs Happiness Score
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='Economy (GDP per Capita)',
    y='Happiness Score',
    data=df,
    color='blue'
)

plt.title('Economy (GDP per Capita) vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')

plt.show()
```

```python
#Health vs Happiness Score
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='Health (Life Expectancy)',
    y='Happiness Score',
    data=df,
    color='green'
)

plt.title('Health (Life Expectancy) vs Happiness Score')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Happiness Score')

plt.show()
```

```python
#Find Most Influential Factors
correlation = df.corr(numeric_only=True)['Happiness Score']

print(correlation.sort_values(ascending=False))
```
